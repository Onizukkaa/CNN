```python
import cv2
import numpy as np
   
cap = cv2.VideoCapture('bassam.mp4') 
   
if (cap.isOpened()== False):  
  print("Error opening video file")

i = 0
while(cap.isOpened()): 
  ret, frame = cap.read() 
  if ret == True:
    cv2.imshow('Frame', frame) 
    i = i + 1
    
    if cv2.waitKey(25) & 0xFF == ord('q'): 
      break
  
  else:  
    break
   
cap.release() 
   
cv2.destroyAllWindows() 
print(i)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
   
cap = cv2.VideoCapture('bassam.mp4') 
   
if (cap.isOpened()== False):  
  print("Error opening video file")

i = 0
j = 1
while(cap.isOpened()): 
  ret, frame = cap.read() 
  if ret == True:
    cv2.imshow('Frame', frame) 
    i = i + 1
    if i == 35*j :
        print(i)
        j += 1
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(image)
        plt.show()

    if cv2.waitKey(25) & 0xFF == ord('q'): 
      break
  
  else:  
    break
   
cap.release() 
   
cv2.destroyAllWindows() 
print(i)
```
